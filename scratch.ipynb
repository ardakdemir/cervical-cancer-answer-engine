```python
import pandas as pd
df = pd.read_csv("data/openalex_cervical_cancer_all_years/OpenAlex_cervical_cancer_screening_2015.csv")
```

```python
file_path = "data/openalex_cervical_cancer_all_years/OpenAlex_cervical_cancer_screening_2015.csv"
```

```python
import csv 
with open(file_path, mode="r", encoding="utf-8",errors='ignore') as f:
    reader = csv.DictReader(f)  # Reads rows as dictionaries
    data = [row for row in reader]  # Convert to list of dicts
    df = pd.DataFrame.from_records(data)  # Convert to DataFrame
```

```python
len(df)
```

```python
import gzip
import pickle as pkl
PARSED_ARTICLES_PROCESSED_SAVE_PATH = "data/parsed_articles_processed.pklz"
with gzip.open(PARSED_ARTICLES_PROCESSED_SAVE_PATH, "rb") as f:
    all_articles = pkl.load(f)
```

```python
len(all_articles)
abstract_lengths = [sum([len(sentence) for sentence in article.article_contents]) for article in all_articles]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Create histogram using seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=abstract_lengths, bins=500)
plt.xlim(0, 200)
plt.title('Distribution of Abstract Lengths')
plt.xlabel('Length of Abstract (characters)')
plt.ylabel('Count')
plt.show()
```
